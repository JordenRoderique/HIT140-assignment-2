```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df1 = pd.read_csv("clean_dataset1.csv")
df2 = pd.read_csv("clean_dataset2.csv")
```

```python
df = df2[['rat_minutes', 'bat_landing_number']].dropna()

plt.figure()
plt.scatter(df['rat_minutes'], df['bat_landing_number'])
plt.xlabel('Rat minutes (per 30-min interval)')
plt.ylabel('Bat landing number')
plt.title('Bat landings vs rat minutes')

z = np.polyfit(df['rat_minutes'], df['bat_landing_number'], 1)
xp = np.linspace(df['rat_minutes'].min(), df['rat_minutes'].max(), 100)
plt.plot(xp, z[0]*xp + z[1])

plt.show()
```

The above shows bat numbers visiting the pad tend to decrease as rats spend more time on the pad. This shows they seem perceive them as potential predators as they are hesitant to approach.

```python
df = df2[['rat_arrival_number', 'bat_landing_number']].dropna()

plt.figure()
plt.scatter(df['rat_arrival_number'], df['bat_landing_number'])
plt.xlabel('Number of rats over 30 minutes')
plt.ylabel('Bat landing number')
plt.title('Bat landings vs Number of rats')

z = np.polyfit(df['rat_arrival_number'], df['bat_landing_number'], 1)
xp = np.linspace(df['rat_arrival_number'].min(), df['rat_arrival_number'].max(), 100)
plt.plot(xp, z[0]*xp + z[1])

plt.show()
```

A similar pattern to above can be seen with amount of bat landings and number of rats visiting the platform.
